```python
nomes = list(jogadores.keys())

def start():
    print('''Neymar, Alisson, Coutinho, Casemiro, Vini Jr, Daniel Alves, Thiago Silva, Lucas Paqueta, Fabinho, Marquinhos\n\nBem-vindo ao jogo!! Pense em um desses jogadores e eu descobrirei em quem você pensou.''')
    x = input('Já pensou no seu jogador? [S/N] ')
    if x.lower().startswith('s'):
        print('''\nEntão vamos lá...\nPor favor, responda as perguntas abaixo apenas com S ou N:\n''')
        question = ['Tem mais de 29 anos? ', 'Jogou / já jogou no Barcelona? ','É atacante? ', 'Jogou na Copa de 2018? ', 'Jogou em um dos grandes do RJ? (Fla, Flu, Vasco, Botafogo) ','Atualmente está Premier League? ']

        teste_dict = {} #dicionario vazio para guardar resultados das perguntas

        r1 = input(question[0]) #Tem mais de 29 anos?
        if r1.lower() == 's': #Caso sim, adicionando jogadores com MAIS DE 29 anos no dicionario
            for age in range(len(nomes)):
                if jogadores[nomes[age]]['Idade']>29:
                    over_29 = []
                    over_29.append(nomes[age])
                    teste_dict[nomes[age]] = jogadores[nomes[age]]
        elif r1.lower() == 'n': #Caso nao, adicionando jogadores com MENOS DE 29 anos no dicionario
            for age in range(len(nomes)):
                if jogadores[nomes[age]]['Idade']<=29:
                    under_29 = []
                    under_29.append(nomes[age])
                    teste_dict[nomes[age]] = jogadores[nomes[age]]

        nomes_r2 = list(teste_dict.keys()) #transformar jogadores filtrados na Q1 para lista

        r2 = input(question[1]) #Jogou / já jogou no Barcelona?
        if r2.lower() == 's': #Caso sim, REMOVER jogadores que NÃO JOGARAM no Barcelona
            for barcelona in range(len(nomes_r2)):
                if teste_dict[nomes_r2[barcelona]]['played_barça'] is False:
                    no_barca = []
                    no_barca.append(nomes_r2[barcelona])
                    for players in range(len(no_barca)):
                        teste_dict.pop(no_barca[players])
        elif r2.lower() == 'n': #Caso nao, REMOVER jogadores que JOGARAM no Barcelona
            for barcelona in range(len(nomes_r2)):
                if teste_dict[nomes_r2[barcelona]]['played_barça'] is True:
                    yes_barca = []
                    yes_barca.append(nomes_r2[barcelona])
                    for players in range(len(yes_barca)):
                        teste_dict.pop(yes_barca[players])

        nomes_r3 = list(teste_dict.keys()) #transformar jogadores filtrados na Q2 para lista

        r3 = input(question[2]) #É atacante?
        if r3.lower() == 's': #Caso sim, REMOVER jogadores que NÃO SÃO atacantes 
            for position in range(len(nomes_r3)):
                if teste_dict[nomes_r3[position]]['Posição'] != 'Atacante':
                    not_forward = []
                    not_forward.append(nomes_r3[position])
                    for players in range(len(not_forward)):
                        teste_dict.pop(not_forward[players])
        elif r3.lower() == 'n': #Caso sim, REMOVER jogadores que SÃO atacantes 
            for position in range(len(nomes_r3)):
                if teste_dict[nomes_r3[position]]['Posição'] == 'Atacante':
                    is_forward = []
                    is_forward.append(nomes_r3[position])
                    for players in range(len(is_forward)):
                        teste_dict.pop(is_forward[players])

        nomes_r4 = list(teste_dict.keys()) #transformar jogadores filtrados na Q3 para lista

        r4 = input(question[3]) #Jogou na Copa de 2018?
        if r4.lower() == 's': #Caso sim, REMOVER jogadores que NÃO JOGARAM a copa de 2018
            for cup2018 in range(len(nomes_r4)):
                if teste_dict[nomes_r4[cup2018]]['copa_2018'] is False:
                    no_2018 = []
                    no_2018.append(nomes_r4[cup2018])
                    for cup in range(len(no_2018)):
                        teste_dict.pop(no_2018[cup])
        elif r4.lower() == 'n': #Caso sim, REMOVER jogadores que JOGARAM a copa de 2018
            for cup2018 in range(len(nomes_r4)):
                if teste_dict[nomes_r4[cup2018]]['copa_2018'] is True:
                    yes_2018 = []
                    yes_2018.append(nomes_r4[cup2018])
                    for cup in range(len(yes_2018)):
                        teste_dict.pop(yes_2018[cup])               

        nomes_r5 = list(teste_dict.keys()) #transformar jogadores filtrados na Q4 para lista               

        r5 = input(question[4]) #Jogou em um dos grandes do RJ? (Fla, Flu, Vasco, Botafogo)
        if r5.lower() == 's': #Caso sim, REMOVER jogadores que NÃO JOGARAM nos grandes de RJ
            for teams_rj in range(len(nomes_r5)):
                if teste_dict[nomes_r5[teams_rj]]['played_grandesrj'] is False:
                    no_rj = []
                    no_rj.append(nomes_r5[teams_rj])
                    for teams in range(len(no_rj)):
                        teste_dict.pop(no_rj[teams])
        elif r5.lower() == 'n': #Caso sim, REMOVER jogadores que JOGARAM nos grandes de RJ
            for teams_rj in range(len(nomes_r5)):
                if teste_dict[nomes_r5[teams_rj]]['played_grandesrj'] is True:
                    yes_rj = []
                    yes_rj.append(nomes_r5[teams_rj])
                    for teams in range(len(yes_rj)):
                        teste_dict.pop(yes_rj[teams])

        nomes_r6 = list(teste_dict.keys()) #transformar jogadores filtrados na Q5 para lista               

        r6 = input(question[5]) #Atualmente está Premier League?
        if r6.lower() == 's': #Caso sim, REMOVER jogadores que NÃO ESTÃO na PL
            for plays_pl in range(len(nomes_r6)):
                if teste_dict[nomes_r6[plays_pl]]['premier_league'] is False:
                    no_pl = []
                    no_pl.append(nomes_r6[plays_pl])
                    for pl in range(len(no_pl)):
                        teste_dict.pop(no_pl[pl])
        elif r6.lower() == 'n': #Caso sim, REMOVER jogadores que ESTÃO na PL
            for plays_pl in range(len(nomes_r6)):
                if teste_dict[nomes_r6[plays_pl]]['premier_league'] is True:
                    yes_pl = []
                    yes_pl.append(nomes_r6[plays_pl])
                    for pl in range(len(yes_pl)):
                        teste_dict.pop(yes_pl[pl])
        
        
        if len(list(teste_dict.keys())) == 1: #Retornar resultado quando restar apenas um na lista
            return 'Seu jogador é o:', list(teste_dict.keys())[0]
        
        else: #Quando inexistente, alertar que houve resposta errada do user.
            print('\nNão consegui identificar seu jogador.\nTalvez você tenha respondido alguma pergunta de forma errada...\n')
            input('Vamos tentar novamente? [S/N]' )
            if x.lower().startswith('s'):
                start()
            else:
                exit()
            
    elif x.lower().startswith('n'): #Se o user ainda não escolheu, pedir para que escolha um jogador
        print('''\nPor favor, escolha seu jogador.\n''')
        start()
    else: #Caso a resposta não seja S ou N, avisar o user
        print('''\nPor favor, responda apenas com S ou N!!\n''')
        start()
```
